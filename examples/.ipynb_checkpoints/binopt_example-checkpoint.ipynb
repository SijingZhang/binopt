```python
rng = np.random.RandomState(15)  # deterministic random data
s = sp.expit(rng.normal(loc= +4, scale=3.0,size=10000)) 
b = sp.expit(rng.normal(loc= -6, scale=3.0,size=50000)) 
ws = np.ones(10000)
wb = np.ones(50000)

plt.figure(figsize=(5,5))

plt.hist(s, bins=500, range=[0,1], weights=ws,
         color='red' ,histtype='step',lw=1.2, normed=0)
plt.hist(b, bins=500, range=[0,1], weights=wb,
         color='blue',histtype='step',lw=1.2, normed=0)

plt.yscale('log')
plt.show()
```

```python
X = np.concatenate([s,b])
Y = np.concatenate([np.ones(s.shape[0]), np.zeros(b.shape[0])])
W = np.concatenate([ws,wb])
```

## checking systematic effect from background

- method A : $Z_A = s / \sqrt(b)$
- method B : $Z_B = s / \sqrt(s + b)$
- method C : $Z_C = s / \sqrt(b + \sigma_b^2)$
- method D : $Z_D = s / \sqrt(s + b + \sigma_b^2)$

```python
from binopt import tools

def significance(tcut, X, Y, W, method="A"):
    s_cdf = np.sum(W[Y==1])*(1 - tools.empirical_cdf(X[Y==1], weights=W[Y==1]).fcn(tcut))
    b_cdf = np.sum(W[Y==0])*(1 - tools.empirical_cdf(X[Y==0], weights=W[Y==0]).fcn(tcut))
    s_err = (1 - tools.empirical_cdf(X[Y==1], weights=W[Y==1]).fcn(tcut))*np.sqrt(np.sum(W[Y==1]**2))
    b_err = (1 - tools.empirical_cdf(X[Y==0], weights=W[Y==0]).fcn(tcut))*np.sqrt(np.sum(W[Y==0]**2))
    
    if method=="A":
        return s_cdf/np.sqrt(b_cdf)
    elif method=="B":
        return s_cdf/np.sqrt(b_cdf + b_cdf)
    elif method=="C":
        return s_cdf/np.sqrt(b_cdf + 1.0/b_err**2)
    elif method=="D":
        return s_cdf/np.sqrt(s_cdf + b_cdf + 1.0/b_err**2)
        
```

```python
plt.figure(figsize=(5,5))
tcut = np.linspace(0.8,1,1000)

plt.plot(tcut, significance(tcut,X,Y,W, "A"), label="$Z_A = s / \\sqrt{b}$")
# plt.plot(tcut, significance(tcut,X,Y,W, "B"), label="$Z_B = s / \\sqrt{s + b}$")
plt.plot(tcut, significance(tcut,X,Y,W, "C"), label="$Z_C = s / \\sqrt{b + \\sigma_b^2}$")
# plt.plot(tcut, significance(tcut,X,Y,W, "D"), label="$Z_D = s / \\sqrt{s + b + \\sigma_b^2}$")

# plt.ylim([0,500])
plt.legend()
# plt.yscale('log')
plt.show()
```

```python
X = np.concatenate([sp.expit(rng.normal(loc= +4, scale=3.0,size=10000)) ,
                    sp.expit(rng.normal(loc= -6, scale=3.0,size=10000)) ])
Y = np.concatenate([np.ones(10000), np.zeros(10000)])
W = np.concatenate([np.ones(10000), np.ones(10000)])
```

```python
def Z_A(s,b):
    rad = (s+b)*np.log(1+(s/b))-s
    return np.sqrt(2*rad)

Z_A = np.vectorize(Z_A)

def Z_B(s,b,berr):
    term_a = (s+b)*(b+berr**2)/(((s+b)*berr**2)+b**2)
    term_b = 1 + (s*berr**2)/(b*(b+berr**2))
    rad = (s+b)*np.log(term_a)-((b**2)/(berr**2))*np.log(term_b)
    return np.sqrt(2*rad)

Z_B = np.vectorize(Z_B)
```

```python
plt.figure(figsize=(6,4))
tcut = np.linspace(0,1,1000)

s_cdf = np.sum(W[Y==1])*(1 - tools.empirical_cdf(X[Y==1], weights=W[Y==1]).fcn(tcut))
b_cdf = np.sum(W[Y==0])*(1 - tools.empirical_cdf(X[Y==0], weights=W[Y==0]).fcn(tcut))
b_err = 1.0/np.sqrt(b_cdf)

# plt.plot(tcut, b_cdf, label="signal")
# plt.plot(tcut, b_err, label="background")
plt.plot(tcut,Z_A(s_cdf,b_cdf), label="$Z_A$")
plt.plot(tcut,Z_B(s_cdf,b_cdf,b_err), label="$Z_B$")
plt.plot(tcut,Z_B(s_cdf,b_cdf,0.1*b_cdf), label="$Z_C$")
# plt.plot(tcut,s_cdf/np.sqrt(b_cdf), '--',label="$s/\\sqrt{b}$")
# plt.plot(tcut,s_cdf/np.sqrt(b_cdf+b_err**2), '--', label="$s/\\sqrt{b+\\sigma_b^2}$")

# plt.ylim([100,10000])
plt.legend(loc='upper left')
# plt.yscale('log')
plt.show()
```

```python
t = np.linspace(0,10,100)
def poisson(k,t):
    return np.exp(k*np.log(t) - t - sp.gammaln(k+1))
poisson= np.vectorize(poisson)

plt.plot(t, st.poisson.pmf(3,t) )
plt.plot(t, poisson(3.2,t) ,'b--')
plt.plot(t, poisson(2.8,t) ,'b--')
```

```python
def divide( a, b ):
    """ ignore / 0, div0( [-1, 0, 1], 0 ) -> [0, 0, 0] """
    with np.errstate(divide='ignore', invalid='ignore'):
        c = np.true_divide( a, b )
        c[ ~ np.isfinite( c )] = 0
    return c

def fom(s, b, berr=0, method="AMS2"):
    """
    Builtin figure of metrits for boundary optimisation.
    4 methods are currently implemented :
    - AMS1 : $Z_0 = s/\sqrt{s + b + \sigma_b^2}$
    - AMS2 : $Z_0 = s/\sqrt{b + \sigma_b^2}$
    - AMS3 : $Z_0 = \sqrt{2(s+b+sigma_b^2)\ln(1 + \frac{s}{b+\sigma_b^2}) - s}$ 
             the $\sigma_b$ should be set to 0 for physical results
    - AMS4 : Significance as defined in [0] but including systematic uncertainty on b    
    [0] Eur. Phys. J. C (2011) 71: 1554 DOI 10.1140/epjc/s10052-011-1554-0
    """
    c = np.zeros(b.shape[0])
    if method == "AMS1":
        c = divide(s , np.sqrt(s + b + berr**2))
    elif method == "AMS2":
        c = divide(s , np.sqrt(b + berr**2 ))
    elif method == "AMS3":
        term_a = (s + b + berr**2)
        term_b = (1 + divide(s, (b + berr**2)))
        c = np.sqrt(2*(term_a * np.log(term_b) - s))
    elif method == "AMS4":
        term_a = divide((s+b)*(b+berr**2),((s+b)*berr**2)+b**2)
        term_b = 1 + divide(s*berr**2, b*(b+berr**2))
        rad = (s+b)*np.log(term_a)-divide(b**2,berr**2)*np.log(term_b)
        c = np.sqrt(2*rad)
    return c

https://inspirehep.net/record/1621076
```
