```python
import matplotlib.pyplot as plt
import numpy  as np
import pandas as pd

%matplotlib inline
plt.style.use('physics')

plt.rcParams['axes.grid'       ]  = False
plt.rcParams['xtick.labelsize' ]  = 14
plt.rcParams['ytick.labelsize' ]  = 14
plt.rcParams['axes.labelsize'  ]  = 14
plt.rcParams['legend.fancybox' ]  = False

pd.options.mode.chained_assignment = None

import binopt

from scipy import special as sp
```

```python
rng = np.random.RandomState(15)  # deterministic random data
s = sp.expit(rng.normal(loc= +4, scale=3.0,size=10000)) 
b = sp.expit(rng.normal(loc= -2, scale=2.9,size=10000)) 
ws = rng.normal(loc= 1 , scale=1 ,size=10000)
wb = rng.normal(loc= 500, scale=1,size=10000)

plt.figure(figsize=(5,5))

plt.hist(s, bins=500, range=[0,1], weights=ws,
         color='red' ,histtype='step',lw=1.2, normed=0)
plt.hist(b, bins=500, range=[0,1], weights=wb,
         color='blue',histtype='step',lw=1.2, normed=0)

plt.yscale('log')
plt.show()
```

```python
X = np.concatenate([s,b])
Y = np.concatenate([np.ones(s.shape[0]), np.zeros(b.shape[0])])
W = np.concatenate([ws,wb])
```

```python
help(binopt.optimize_bin)
```

```python
binner = binopt.optimize_bin(nbins=4, range=[0,1], drop_last_bin=False, 
                             fix_upper=True, fix_lower=False, 
                             use_kde_density=False)

bin_dl = binopt.optimize_bin(nbins=4, range=[0,1], drop_last_bin=True, 
                             fix_upper=True, fix_lower=False, 
                             use_kde_density=False)
```

```python
help(binner.fit)
```

```python
min_args={
    'maxiter' : 10000
}
binner.fit(X,Y, sample_weights=W, niter=10, min_args=min_args)
bin_dl.fit(X,Y, sample_weights=W, niter=10, min_args=min_args)
```

```python
fig = plt.figure(figsize=(12,5))
binner.optimisation_monitoring_(fig)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
bin_dl.optimisation_monitoring_(fig)
plt.show()
```

```python
print binner.result.x
print bin_dl.result.x
print binner.x_init
```

```python
plt.figure(figsize=(5,5))

plt.hist(s, bins=500, range=[0,1], weights=ws,
         color='red' ,histtype='step',lw=1.2, normed=0)
plt.hist(b, bins=500, range=[0,1], weights=wb,
         color='blue',histtype='step',lw=1.2, normed=0)

for x in binner.result.x:
    plt.axvline(x, ls='--', c = 'g',label='drop-last-bin')
    
for x in bin_dl.result.x:
    plt.axvline(x, ls='-', c = 'r', label='inclusive')
    

plt.yscale('log')
plt.xlim([0,1])
plt.legend()
plt.show()
```

```python
from iminuit import Minuit
# 0.    0.25  0.5   0.75  1.
def minuit_cost(a, b, c, d):
    return binner.cost_fun(np.array([a,b,c,d]))
m = Minuit(minuit_cost, 
           a=0,b=0.3,c=0.6,d=1, 
           limit_b=(0,1), 
           limit_c=(0,1),
           error_b=0.01,
           error_c=0.01,
           fix_a=True, fix_d=True)
m.migrad()
```

```python
print dir(m)
m.values
```

```python
plt.figure(figsize=(5,5))

plt.hist(s, bins=500, range=[0,1], weights=ws,
         color='red' ,histtype='step',lw=1.2, normed=0)
plt.hist(b, bins=500, range=[0,1], weights=wb,
         color='blue',histtype='step',lw=1.2, normed=0)

for i,x in m.values.items():
    plt.axvline(x, ls='--',c='k',label=i)

plt.yscale('log')
plt.xlim([-0.1,1.1])
plt.legend()
plt.show()
```
