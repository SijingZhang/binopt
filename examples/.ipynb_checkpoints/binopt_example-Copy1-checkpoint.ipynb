```python
def divide( a, b ):
    """ ignore / 0, div0( [-1, 0, 1], 0 ) -> [0, 0, 0] """
    with np.errstate(divide='ignore', invalid='ignore'):
        c = np.true_divide( a, b )
        c[ ~ np.isfinite( c )] = 0
    return c

def fom(s, b, berr=0, method="AMS2"):
    """
    Builtin figure of metrits for boundary optimisation.
    4 methods are currently implemented :
    - AMS1 : $Z_0 = s/\sqrt{s + b + \sigma_b^2}$
    - AMS2 : $Z_0 = s/\sqrt{b + \sigma_b^2}$
    - AMS3 : $Z_0 = \sqrt{2(s+b+sigma_b^2)\ln(1 + \frac{s}{b+\sigma_b^2}) - s}$ 
             the $\sigma_b$ should be set to 0 for physical results
    - AMS4 : Significance as defined in [0] but including systematic uncertainty on b    
    [0] Eur. Phys. J. C (2011) 71: 1554 DOI 10.1140/epjc/s10052-011-1554-0
    """
    c = np.zeros(b.shape[0])
    if method == "AMS1":
        c = divide(s , np.sqrt(s + b + berr**2))
    elif method == "AMS2":
        c = divide(s , np.sqrt(b + berr**2 ))
    elif method == "AMS3":
        term_a = (s + b + berr**2)
        term_b = (1 + divide(s, (b + berr**2)))
        c = np.sqrt(2*(term_a * np.log(term_b) - s))
    elif method == "AMS4":
        term_a = divide((s+b)*(b+berr**2),((s+b)*berr**2)+b**2)
        term_b = 1 + divide(s*berr**2, b*(b+berr**2))
        rad = (s+b)*np.log(term_a)-divide(b**2,berr**2)*np.log(term_b)
        c = np.sqrt(2*rad)
    return c

def mass_sigma_effective(X, Y, W, mass, bins, nsig=1):
    """
    Input should contain a resonance of some sort.
    """
    _cats_ = np.digitize(X, bins)
    _sige_ = []
    _cats_ = np.digitize(X, bins)
    _sige_ = []
    print np.unique(_cats_)
    for cid in range(bounds.shape[0]):
        min_, max_ = tools.weighted_quantile(
            mass[_cats_ == cid],
            [norm.cdf(0, -nsig, 1), norm.cdf(0, nsig, 1)],
            sample_weight=W[_cats_ == cid])
        _sige_.append(np.abs(max_-min_)/2.0)
    return np.array(_sige_)
```

## LLH fit

```python
import scipy.stats as st
from scipy.optimize import minimize
```

```python
def exponotial(x, lamb):
    return lamb * np.exp( -lamb * (x - 100) )

def logpdf(params, data):
    lamb = params
    p2 = st.expon.logpdf(data, loc=100, scale=np.exp(lamb ))
    return p2

def nll(params, data, weight):
    out = -np.sum(logpdf(params, data)*weight)
    return out
```

```python
_data_ = M[(Y==0)&(X>0.5)]
_w_ = W[(Y==0)&(X>0.5)]

def objective(x):
    params = x 
    out = nll(params, _data_, _w_)
    if np.isnan(out):
        raise ValueError("Illegal probability {}".format(params))
    return out
```

```python
out_bkg = minimize(objective, x0=[0.03], method='Powell')
out_bkg
```

### Combine everything

```python
def model(data, params):
    loc, scale = out_sig.x
    lamb = out_bkg.x
    s, b = params
    p1 = st.norm(loc=loc, scale=np.sqrt(scale**2)).pdf(data)
    p2 = st.expon(loc=100, scale=np.exp(lamb )).pdf(data)
    return (s/(s+b))*p1 + (b/(s+b))*p2

def logmodel(params, data):
    loc, scale = out_sig.x
    lamb = out_bkg.x
    s, b= params
    with np.errstate(divide='ignore', invalid='ignore'):
        p1 = np.log(s) + st.norm(loc=loc, scale=np.sqrt(scale**2)).logpdf(data)
        p2 = np.log(b) + st.expon(loc=100, scale=np.exp(lamb)).logpdf(data)
        L = np.logaddexp(p1, p2)
        L[ ~ np.isfinite( L )] = 0
        L[ ~ np.isnan( L )] = 0
    return L

def nll(params, data, weight):
    s, b = params
    out = (s+b) - np.sum(logmodel(params, data)*weight)
    return out

_data_ = M[(X>0.5)]
_w_ = W[(X>0.5)]

def objective(x):
    params = x 
    out = nll(params, _data_, _w_)
    if np.isnan(out):
        raise ValueError("Illegal probability {}".format(params))
    return out
```

```python
def draw_phase_space(X,Y):
    target = np.unique(Y)
    colors_= ['#1f77b4', '#d62728', '#98df8a',  '#ff9896']
    print X.ndim
    if X.ndim < 2: return
    for i in range(X.ndim):
        for j in range(i+1,X.ndim):
            plt.figure(figsize=(5,5))
            for c in target:
                plt.scatter(X[:,i][Y==c], X[:,j][Y==c], 
                            marker='.', s=2.5, alpha=0.5,
                            color=colors_[int(c)])
            plt.xlim([X[:,i][Y==c].min(), X[:,i][Y==c].max()])
            plt.ylim([X[:,j][Y==c].min(), X[:,j][Y==c].max()])
            plt.legend(fontsize=13)
```
